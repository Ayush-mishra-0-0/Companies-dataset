```python
final_df= pd.DataFrame()
for j in range(1,504):
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage= requests.get('https://www.ambitionbox.com/list-of-companies?page={}'.format(j),headers=headers).text
    soup=bs(webpage,'lxml')
    companies= soup.find_all('div',class_='companyCardWrapper')
    name=[]
    rating=[]
    fields = []
    employee_counts = []
    company_types = []
    ages = []
    locations = []

    for i in companies:
        name.append(i.find('h2').text.strip())
        rating.append(i.find('div', class_='companyCardWrapper__companyRating').text.strip())
        data = i.find('div', class_='companyCardWrapper__interLinkingWrapper').text.strip()
        parts = data.split('|')

        # Store specific parts in respective columns, and set None for missing parts
        fields.append(parts[0].strip() if len(parts) > 0 else None)
        employee_counts.append(parts[1].strip() if len(parts) > 1 else None)
        company_types.append(parts[2].strip() if len(parts) > 2 else None)
        ages.append(parts[3].strip() if len(parts) > 3 else None)
        locations.append(parts[4].strip() if len(parts) > 4 else None)

    df = pd.DataFrame({
        "name": name,
        "rating": rating,
        "fields": fields,
        "employee_counts": employee_counts,
        "company_types": company_types,
        "ages": ages,
        "locations": locations
    })
    final_df = pd.concat([final_df, df], ignore_index=True)
```
